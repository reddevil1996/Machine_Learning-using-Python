```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
sb.set()
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidian Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_pred = ac.fit_predict(X)
```

```python
y_pred
```

```python
plt.scatter(X[y_pred == 0, 0], X[y_pred == 0, 1], s = 50, c = 'red', label = 'Careful')
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], s = 50, c = 'blue', label = 'Standard')
plt.scatter(X[y_pred == 2, 0], X[y_pred == 2, 1], s = 50, c = 'green', label = 'Target')
plt.scatter(X[y_pred == 3, 0], X[y_pred == 3, 1], s = 50, c = 'cyan', label = 'Careless')
plt.scatter(X[y_pred == 4, 0], X[y_pred == 4, 1], s = 50, c = 'magenta', label = 'Sensible')
plt.title('Hierarchical Clustering')
plt.xlabel('Income')
plt.ylabel('Spending')
plt.legend()
plt.show()
```
