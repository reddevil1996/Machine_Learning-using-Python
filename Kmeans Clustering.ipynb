```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
sb.set()
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Graph')
plt.xlabel('No of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)
y_pred = kmeans.fit_predict(X)
```

```python
y_pred
```

```python
plt.scatter(X[y_pred == 0, 0], X[y_pred == 0, 1], s = 50, c = 'red', label = 'Careful')
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], s = 50, c = 'blue', label = 'Standard')
plt.scatter(X[y_pred == 2, 0], X[y_pred == 2, 1], s = 50, c = 'green', label = 'Target')
plt.scatter(X[y_pred == 3, 0], X[y_pred == 3, 1], s = 50, c = 'cyan', label = 'Careless')
plt.scatter(X[y_pred == 4, 0], X[y_pred == 4, 1], s = 50, c = 'magenta', label = 'Sensible')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s =100, c = 'black', label = 'Centroid')
plt.title('Kmeans Clustering')
plt.xlabel('Income')
plt.ylabel('Spending')
plt.legend()
plt.show()
```
